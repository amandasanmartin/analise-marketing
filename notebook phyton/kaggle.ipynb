```python
#Conexão com o banco de dados
!pip install sqlalchemy psycopg2

# Importando a ferramenta do motor
from sqlalchemy import create_engine

# 1. Configurando as credenciais
usuario = 'seu_usuario_aqui'
senha = 'sua_senha_aqui' 
host = 'seu_local_host'
porta = 'sua_porta'
banco = 'seu_banco' 

# 2. Montando a string de conexão
string_conexao = f"postgresql://{usuario}:{senha}@{host}:{porta}/{banco}"

# 3. Criando o motor de envio
engine = create_engine(string_conexao)

print("Motor de conexão criado com sucesso!")
```

```python
#Envio para o Banco

print("Iniciando a carga dos dados na camada Staging...")

# 1. Tabela de Leads (Topo do Funil)
df_funil.to_sql('tb_funil_marketing', con=engine, schema='staging', if_exists='replace', index=False)

# 2. Tabela de Conversão (Contratos Fechados)
df_conv.to_sql('tb_negocios_fechados', con=engine, schema='staging', if_exists='replace', index=False)

# 3. Tabela de Faturamento (Itens Vendidos)
df_fat.to_sql('tb_itens_faturados', con=engine, schema='staging', if_exists='replace', index=False)

# 4. Tabela de Status dos Pedidos (Filtro de Qualidade)
df_an.to_sql('tb_status_pedidos', con=engine, schema='staging', if_exists='replace', index=False)

print("Carga completa! Dados organizados no esquema 'staging'.")
```
